```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
```

### Preprocessing/Cleaning

```python
df = pd.read_csv('data.csv').dropna(axis='rows')
df.columns = ['url', 'headline', 'body', 'label']
df.shape
```

```python
df.dtypes
```

### Using only Header

```python
vect = CountVectorizer()
tfid = TfidfTransformer() 
X = vect.fit_transform(df['headline'].astype(str)).copy()
X = tfid.fit_transform(X)
y = df.loc[:, 'label'].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
X_train.shape
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)

assert y_pred.shape == y_test.shape
plot_confusion_matrix(y_test, y_pred, classes=np.array(['fake', 'legit']), normalize=True,
                      title='Normalized confusion matrix');
```
